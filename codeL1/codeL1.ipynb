```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_linear.csv').values #import data set
```

```python
N = data.shape[0] #get the number of row
x = data[:, 0].reshape(-1, 1) #get the first row and reshape to 2D with 1 column
y = data[:, 1].reshape(-1, 1)
plt.scatter(x, y) #draw the graph
plt.xlabel('m2')
plt.ylabel('price')

```

```python
x = np.hstack((np.ones((N, 1)), x)) #make the matrix with N row and 1 column and add to the front of x
```

```python
w = np.array([0.,1.]).reshape(-1, 1) 
```

```python
numOfIteration = 100 #amount of alpha step
cost = np.zeros((numOfIteration, 1))#save the value of cost function
learning_rate = 0.000001 #lenght of alpha step
for i in range(1, numOfIteration):
    r = np.dot(x, w) - y # caculated the distance between real nums and predict
    cost[i] = 0.5*np.sum(r*r) # cost func
    w[0] -= learning_rate*np.sum(r) #update 
    w[1] -= learning_rate*np.sum(np.multiply(r, x[:, 1].reshape(-1, 1))) #update
    print(cost[i])
predict = np.dot(x ,w)
plt.plot((x[0][1], x[N-1][1]),(predict[0], predict[N-1]), 'r')
plt.show()
    
```

```python
x1 = 50
y1 = w[0] + w[1]*x1
print('price of 50 m2 house is : ', y1)
```
